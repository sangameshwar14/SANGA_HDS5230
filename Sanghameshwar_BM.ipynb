```python
import matplotlib.pyplot as plt
import numpy as np

steps = ['Dummy', 'KNN Training', 'KNN Prediction', 'KNN Retraining', 'KNN Retrained Prediction']
accuracy = [0.5075, 0.8983, 0.895, 0.9717, 0.965]
recall = [0.5048, 0.8867, 0.4163, 0.9407, 0.913]
precision = [0.5327, 0.9045, 1.0, 0.9061, 0.8873]
plt.figure(figsize=(12, 6))
x = np.arange(len(steps))
plt.plot(x, accuracy, 'o-', linewidth=2, label='Accuracy', color='blue')
plt.plot(x, recall, 's-', linewidth=2, label='Recall', color='green')
plt.plot(x, precision, '^-', linewidth=2, label='Precision', color='red')
plt.xlabel('Model Evolution Steps')
plt.ylabel('Performance Metric Value')
plt.title('Batch Model Performance Across Evolution Steps')
plt.xticks(x, steps, rotation=15)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
